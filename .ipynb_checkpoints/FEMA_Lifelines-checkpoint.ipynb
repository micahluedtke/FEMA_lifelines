## Hazmat Data Import
Hazmat data imported as geopandas DataFrame. Address latitude and longitude found using geopy and converted to Point objects.

```python
hazmat = gpd.read_file('hazmat.csv')
```

```python
hazmat.head()
```

```python
hazmat['FIPS'] = 0
```

```python
hazmat.head()
```

```python
# Geopy could not locate the original address for row 2, so we looked up and found the following address:
hazmat['Hazardous'][2] = '224 E Main St, West Brookfield, MA 01585'
```

```python
hazmat.drop(columns=['Waste','Addresses'],inplace=True)
```

```python
for i in range(len(hazmat)):
    location = geolocator.geocode(hazmat['Hazardous'][i])
    try:
        coord = float(location.longitude), float(location.latitude)
        coord = tuple(coord)
        hazmat['geometry'][i] = coord 

    except:
        print('We could not location {} in our geopy search'.format(hazmat['Hazardous'][i]))
```

```python
hazmat[['geometry']].head()
```
